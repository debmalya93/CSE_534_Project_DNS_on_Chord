```python
import dns
import dns.resolver
import pandas as pd
import time

df = pd.read_csv('Final-400K_dev_10K.txt', header=None).reset_index(drop=True)
answer_lookup = []
for i in range(0, len(df)):
    website = df.iloc[i][0].split(' ')[0]
    start = time.time()
    try:
        result = dns.resolver.resolve(website, 'A')
    except (dns.resolver.NXDOMAIN,dns.resolver.NoAnswer):
        continue
    end = time.time()
    answer_lookup.append((website, (end - start)*1000))
    print(answer_lookup[-1])    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(answer_lookup, columns = ['website','lookup_time'])
plt.figure(figsize=(25,6))
sns.set(font_scale = 1.2)
plot = sns.lineplot(x="website", y="lookup_time", data=df.head(2))
plot.set_title("Traditional DNS Resolution Lookup Times")
plot.set_yticks(range(0, int(df.lookup_time.max())), 20)
```
